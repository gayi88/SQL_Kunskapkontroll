1.	Describe briefly how a relational database works?  
Ans: It works first by storing data using tables (which contains both rows and columns).Using primary and foreign keys it creates relation between different tables in different ways (Ex:One to One,One to Many,Many to Many). It uses Structured Query Launguage to manupulate data and apply Normalization to reduce unexpected fails.
	
2.  What is meant by CRUD flow?  
Ans: CRUD is a acronym to Create, Read, Update, Delete. The term includes the basic operations that can be performed on data in a relational database.
C : Create or add rows to the table.
R : Read data from a table.
U: Update existing data in a table
D: Removes rows from a table.

3.	Describe briefly what is a ‘Left Join’ and ‘Inner Join’ ? Why do we use them?  
Ans: Left Join:  It retrieves all the rows from left table and also retrieves matching rows from the right. If right table rows does not match with left table, then it includes  all the rows from left table, but the columns from the right table will contain NULL values. We use this join to get all the rows from a specific table that does not have corresponding match in other table.
Inner Join: It retrieves only the rows that have matching rows in both tables. The row will not be retrieved if there is no match. We use when we need to see data present in both the tables.

4. Briefly describe what indexing in SQL meant?  
Ans: Indexing is used to find rows that meet a certain criteria in a specific column. By using Indexing, system can directly search among the values stored in the Index, else system would go row by row to find macthing rows.Common Index types are Clustured Index and Non-Clustered Index.

5. Describe briefly what is View in SQL?  
Ans: It is a virtual table whose contents are defined by a query, which makes data easily accessible in a table format. The complex queries need not to be re-run every time we want to access the data. Due to authorization and security issues every one cannot access all the data, so View allows us to share certain selected information who does have access to the database.

6. Describe briefly what is stored procedure in Sql?  
Ans: It is a saved code that can be re-used when we want to access and can be quicker to call than writing a code .Using stored procedures allows to centralize the definations. Consistent results and common definations are ensured if these are used by multiple people.

 <span style = "color:Purple"> 1. EMPLOYEE LIST BY GENDER , HIRING DATE ,SERVING YEARS</span>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Common Table Expression
cte_HR = """
WITH [EmployeeDetails] AS 
(
    SELECT 
        Person.PersonType,
        Person.FirstName AS [First Name],
        Person.LastName AS [Last Name],
        HumanResources.Employee.NationalIDNumber AS [NationalID Number],
        Employee.Gender AS [Employee Gender],
        FORMAT(Employee.BirthDate, 'dd-MMM-yyyy') AS [Birth Date],
        DATEDIFF(YEAR, BirthDate, GETDATE()) AS [Age of Employee],
        Employee.LoginID,
        Employee.OrganizationLevel,
        Shift.ShiftID,
        CONCAT(Shift.StartTime, ' - ', Shift.EndTime) AS [Shift Timings],
        Shift.Name AS [Shift Name],
        Employee.JobTitle,
        Department.Name AS [Department Name],
        Department.GroupName AS [Functional Group],
        Employee.HireDate AS [Hire Date]
    FROM 
        HumanResources.Employee
    JOIN 
        Person.Person ON HumanResources.Employee.BusinessEntityID = Person.Person.BusinessEntityID
    JOIN 
        HumanResources.EmployeeDepartmentHistory ON Person.Person.BusinessEntityID = HumanResources.EmployeeDepartmentHistory.BusinessEntityID
    JOIN 
        HumanResources.Department ON HumanResources.Department.DepartmentID = HumanResources.EmployeeDepartmentHistory.DepartmentID
    JOIN 
        HumanResources.Shift ON HumanResources.Shift.ShiftID = HumanResources.EmployeeDepartmentHistory.ShiftID
)
"""

# JOINING COMMON TABLE EXPRESSION WITH NEW QUERY TO CHECK NUMBER OF EMPLOYEES BASED ON THEIR GENDER.
query1 = cte_HR + """
SELECT 
    [Employee Gender], 
    COUNT(*) AS [Number of Employees]
FROM 
    EmployeeDetails
GROUP BY 
    [Employee Gender]
ORDER BY 
    [Number of Employees] DESC;
"""

 # JOINING COMMON TABLE EXPRESSION WITH NEW QUERY TO CHECK PEOPLE HIRED IN RESPECTIVE YEARS
query2 = cte_HR + """
SELECT 
    YEAR([Hire Date]) AS [Year of hire],
    COUNT(*) AS [New hires]
FROM 
    EmployeeDetails
GROUP BY 
    YEAR([Hire Date])
ORDER BY 
    [Year of hire] DESC;
"""

# JOINING COMMON TABLE EXPRESSION WITH NEW QUERY TO CHECK SERVING YEARS OF EACH EMPLOYEE 
query3 = cte_HR + """
SELECT 
    [First Name],
    [Last Name],
    [Hire Date],
    DATEDIFF(YEAR, [Hire Date], GETDATE()) AS [Serving Years]
FROM 
    EmployeeDetails;
"""

# EXECUTE ALL QUERIES
with engine.connect() as connection:
    df_gender = pd.read_sql(query1, connection)
    df_hires = pd.read_sql(query2, connection)
    df_retention = pd.read_sql(query3, connection)

# DISPLAY RESULTS
print("Gender Analysis:")
print(df_gender)

print("\nHire Date Analysis:")
print(df_hires)

print("\nRetention Analysis:")
print(df_retention)


#Visualization
#1.
plt.bar(df_gender['Employee Gender'],df_gender['Number of Employees'], color = 'orange')
plt.title('Employee Gender Distribution')
plt.xlabel('Employee Gender')
plt.ylabel('Number of Employees')

plt.tight_layout()
plt.show()

#2.

plt.plot(df_hires['Year of hire'],df_hires['New hires'], color = 'orange')
plt.title('Hiring Data')
plt.xlabel('Year of hire')
plt.ylabel('New hires')
plt.show()

#3.

plt.hist(df_retention['Serving Years'], bins=range(0, df_retention['Serving Years'].max()), color='orange', edgecolor='black')
plt.title('Employee service years')
plt.xlabel('Years of service')
plt.ylabel('No of Employees')
plt.show()

```

<span style = "color:purple">2. CUSTOMERS BY CITY, CUTOMERS BY COUNTRY,CUSTOMERS BY CONTINENT </span>

```python
import pandas as pd
import matplotlib.pyplot as plt


# Common Table Expression
cte_Customer = """
WITH [Customer Analysis] AS 
(
Select Concat(person.FirstName, ' ', person.LastName) As [Customer Name]
        ,Address.City As [City Name]
        ,SalesTerritory.Name As [Country Name]
        ,SalesTerritory.[group] As [Continent Name]
        ,COUNT(SalesOrderHeader.SalesOrderID) As [Number of Orders]
        ,SUM(SalesOrderHeader.SalesOrderID) As [Sales Lastyear]
        ,Ceiling(SUM(SalesOrderHeader.TotalDue)) As [Sales Ytd]
    
from Person.Person 
join Sales.Customer 
    on Sales.Customer.PersonID = person.person.BusinessEntityID 
join Sales.SalesOrderHeader 
    on Sales.Customer.CustomerID = Sales.SalesOrderHeader.CustomerID
join Sales.SalesTerritory 
    on Sales.SalesOrderHeader.TerritoryID = Sales.SalesTerritory.TerritoryID
left join Person.BusinessEntityAddress
    on Sales.Customer.CustomerID = BusinessEntityAddress.BusinessEntityID 
join Person.Address
    on BusinessEntityAddress.AddressID = Person.Address.AddressID
where PersonType ='IN'
GROUP BY 
      Person.FirstName, Person.LastName, Address.City, SalesTerritory.Name, 
      SalesTerritory.[Group], Sales.SalesOrderHeader.TotalDue,
      Sales.SalesOrderHeader.SalesOrderID)
"""
# Customer count by City Name
query1 = cte_Customer + """
Select [City Name],
        COUNT(*) as [Number Of Customers]
from [Customer Analysis]
where [City Name] IS NOT NULL
Group by [City Name];
"""
# Customer count by Country Name
query2 = cte_Customer + """
select [Country Name],
        COUNT(*) As [Number Of Customers]
from [Customer Analysis]
Group by [Country Name]
Order by [Number Of Customers] Desc;
"""
# Customer count by Continent Name
query3 = cte_Customer + """
Select [Continent Name],
        COUNT (*) As [Number of Customers]
from [Customer Analysis]
GROUP BY 
    [Continent Name]
ORDER BY 
    [Continent Name];
"""

# EXECUTE ALL QUERIES
with engine.connect() as connection:
    df_City = pd.read_sql(query1, connection)
    df_Country = pd.read_sql(query2, connection)
    df_Continent = pd.read_sql(query3, connection)
  

# DISPLAY RESULTS
print("City:")
print(df_City)

print("\nCountry:")
print(df_Country)

print("\nContinent:")
print(df_Continent)

#Vizualisation of results
plt.figure(figsize=(8, 6))
plt.bar(df_Country['Country Name'], df_Country['Number Of Customers'], color='orange')
plt.xlabel('Country Name', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.title('Customer Count by Country', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#Vizualisation of results
plt.figure(figsize=(10, 6))
plt.bar(df_Continent['Continent Name'], df_Continent['Number of Customers'], color='green')
plt.xlabel('Continent Name', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.title('Customer Count by Continent',fontsize =14)
plt.tight_layout()
plt.show()
```

<span style = "color:purple">3. 'PROFIT'- BASED ON EACH CATEGORY NAME </span>

```python

query = """
SELECT 
    ProductCategory.ProductCategoryID,
    ProductCategory.Name AS [Product Category Name],
    SUM(OrderDetails.TotalOrderQty) AS [Total Order Quantity],
    SUM(OrderDetails.TotalStandardCost) AS [Total Standard Cost],
    CEILING(SUM(OrderDetails.TotalLineTotal)) AS [Total Sales Amount],
    CEILING(SUM(OrderDetails.TotalLineTotal) - SUM(OrderDetails.TotalStandardCost)) AS [Total Profit] -- Profit Calculation
FROM 
    Production.ProductCategory 
JOIN 
    Production.ProductSubcategory ProductSubcategory 
    ON ProductCategory.ProductCategoryID = ProductSubcategory.ProductCategoryID
JOIN 
    Production.Product 
    ON ProductSubcategory.ProductSubcategoryID = Product.ProductSubcategoryID
JOIN 
    (   -- Subquery to calculate Total Order Quantity, Total Line Total, and Total Standard Cost
        SELECT 
            SalesOrderDetail.ProductID,
            SUM(SalesOrderDetail.OrderQty) AS TotalOrderQty,
            SUM(SalesOrderDetail.LineTotal) AS TotalLineTotal,
            SUM(SalesOrderDetail.OrderQty * COALESCE(Product.StandardCost, 0)) AS TotalStandardCost -- Total Standard Cost
        FROM 
            Sales.SalesOrderDetail SalesOrderDetail
        JOIN 
            Production.Product Product
            ON SalesOrderDetail.ProductID = Product.ProductID
        GROUP BY 
            SalesOrderDetail.ProductID
    ) AS OrderDetails 
    ON Product.ProductID = OrderDetails.ProductID
GROUP BY 
    ProductCategory.ProductCategoryID,
    ProductCategory.Name
ORDER BY 
    ProductCategory.ProductCategoryID ASC;
"""

# Execute the query and load the data into a DataFrame
with engine.connect() as connection:
    df = pd.read_sql(query, connection)

# Print the DataFrame
print(df)

#Visualize Results

plt.figure(figsize=(8,6))
plt.barh(df['Product Category Name'],df['Total Profit'],color ='purple')
plt.xlabel('Total profit')
plt.ylabel('Product Category')
plt.title('Profit-by product category')
plt.tight_layout()
plt.show()
```

<span style = "color:purple">4. DISTRIBUTION OF SALES BASED ON TYPES OF REASONS </span>

```python
# People purchased products based on different parameters
# We retrieve which parameter attracted more customers


query = """
SELECT 
    SalesReason.Name AS [Reason Type],
    SUM(SalesOrderDetail.OrderQty) AS TotalSalesQuantity
FROM 
    Sales.SalesReason
JOIN 
    Sales.SalesOrderHeaderSalesReason
    ON Sales.SalesOrderHeaderSalesReason.SalesReasonID = Sales.SalesReason.SalesReasonID
JOIN 
    Sales.SalesOrderDetail
    ON Sales.SalesOrderDetail.SalesOrderID = Sales.SalesOrderHeaderSalesReason.SalesOrderID
GROUP BY 
    SalesReason.Name
ORDER BY 
    TotalSalesQuantity DESC;

"""

# Execute the query and load the data into a DataFrame
with engine.connect() as connection:
    df = pd.read_sql(query, connection)

# Print the DataFrame
print(df)


#Vizualize results
plt.figure(figsize=(6,6))
plt.pie(df['TotalSalesQuantity'],labels=df['Reason Type'])
plt.title('Proportion of Total Sales Quantity by sales type')
plt.show()

```

<span style = "color:red"> STATISTICAL ANALYSIS </span>

```python

from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
# Execute the query and load the data into a DataFrame
query = """
SELECT 
    ProductCategory.Name AS [Product Category Name],
    sum(FLOOR(SalesOrderDetail.OrderQty * SalesOrderDetail.LineTotal)) AS [Total Revenue]
FROM 
    Sales.SalesOrderDetail
JOIN 
    Production.Product
    ON Sales.SalesOrderDetail.ProductID = Production.Product.ProductID
JOIN 
    Production.ProductSubcategory 
    ON Production.Product.ProductSubcategoryID = Production.ProductSubcategory.ProductSubcategoryID
JOIN 
    Production.ProductCategory
    ON Production.ProductSubcategory.ProductCategoryID = Production.ProductCategory.ProductCategoryID
JOIN 
    Sales.SalesOrderHeader 
    ON Sales.SalesOrderDetail.SalesOrderID = Sales.SalesOrderHeader.SalesOrderID

WHERE YEAR(SalesOrderHeader.OrderDate) = 2014
group by ProductCategory.Name
ORDER BY 
    [Total Revenue] DESC
;
"""

with engine.connect() as connection:
    df = pd.read_sql(query, connection)


total_revenue = df['Total Revenue'].sum() #Total revenue for categories


df['Revenue Proportion'] = df['Total Revenue'] / total_revenue #proportion for each product category

print(df[['Product Category Name', 'Total Revenue', 'Revenue Proportion']]) #print proportions


# Visualisation using pie chart
plt.figure(figsize=(10, 6))
plt.pie(df['Revenue Proportion'], labels=df['Product Category Name'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Total Revenue by Product Category (2014)')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is a circle.
plt.show()

```

```python
import numpy as np
import pandas as pd

#Calculate Total revenue

total_revenue = df['Total Revenue'].sum()

#Proportions for product categories
Product_category = {
    'Bikes': 0.842409,
    'Components': 0.090599,
    'Clothing': 0.045202,
    'Accessories': 0.021789
}

Z_score = 1.96 # z-score 95% CI
n = len(Product_category) 

#Calculate CI
def calc_confidence_interval(name, p, n, z):
   
    se = np.sqrt((p * (1 - p)) / n) #standard error
    margin_of_error = z * se #Margin of error

  
    ci_lower = max (0,p - margin_of_error) #lower bound CI
    ci_upper = p + margin_of_error # Upper bound C

    return name, p, ci_lower, ci_upper
results = [] #for storing output

for category, proportion in Product_category.items():
    name, p, ci_lower, ci_upper = calc_confidence_interval(category, proportion, n, Z_score)
    results.append([name, p, ci_lower, ci_upper])


df_results = pd.DataFrame(results, columns=['Product Category', 'Proportion', 'CI Lower', 'CI Upper']) #dataframe for results
print(df_results) #Print the results
```
